Caption EXCEL 수정

```python
import pandas as pd

captions = pd.read_excel('captions.xlsx')
```

```python
captions.head()
```

```python
captions = captions.iloc[:,3:]
captions.head()
```

```python
import re

captions['head'] = captions['route'].apply(lambda x: re.search('\D*(?=\d)',x.split('/')[-1]).group())
captions['name'] = 0
for i in range(len(captions)):
    x = captions.iloc[i]
    if len(x.route.split('/'))==1:
        captions.loc[i,'name'] = f'{x.route}' 
    else:
        k = x.route.split('/')[-1]
        captions.loc[i,'name'] = f'{k}'
captions.head()
```

```python
captions.to_csv('captions.csv',index=False)
```
